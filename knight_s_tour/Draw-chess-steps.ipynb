```python
def idx2loc(idx, nx):
    return (idx-0.5)/nx

def draw_arrows(ax, locs, nx, ny):
    ax.plot([idx2loc(locs[0][0], nx)], [idx2loc(locs[0][1], ny)],
            markersize=15, marker='o', markerfacecolor='green')
    for i in xrange(len(locs)-1):
        loc = locs[i]
        locc = locs[i+1]
        x = idx2loc(loc[0], nx)
        y = idx2loc(loc[1], ny)
        dx = idx2loc(locc[0], nx) - idx2loc(loc[0], nx)
        dy = idx2loc(locc[1], ny) - idx2loc(loc[1], ny)
        ax.arrow(x, y, dx, dy, color='red',
                 length_includes_head=True, width=0.001, head_width=0.01, overhang=0.5)
    ax.plot([idx2loc(locs[-1][0], nx)], [idx2loc(locs[-1][1], ny)],
            markersize=15, marker='o', markerfacecolor='blue')
    return


def draw_chess_trace(locs, nx=None, ny=None, size=0.7, close=False):
    c1 = (0,0,0)
    c2 = (0.7,0.7,0.7)
    sz = size
    if nx == None or ny == None:
        nx = max(loc[0] for loc in locs)
        ny = max(loc[1] for loc in locs)
    f = figure(figsize=(nx * sz, ny * sz))
    ax = f.add_axes([0,0,1,1])
    ax.set_xticks([])
    ax.set_yticks([])
    dx = 1.0 / nx
    dy = 1.0 / ny
    for i in xrange(nx):
        x = i*dx
        for j in xrange(ny):
            y = j*dy
            color = c1 if (i+j)%2 == 0 else c2
            ax.add_patch(Rectangle((x, y), dx, dy, color=color))
    draw_arrows(ax, locs, nx, ny)
    if close:
        plot([idx2loc(locs[0][0], nx), idx2loc(locs[-1][0], nx)],
             [idx2loc(locs[0][1], ny), idx2loc(locs[-1][1], ny)], color='magenta')
    
    return
```
